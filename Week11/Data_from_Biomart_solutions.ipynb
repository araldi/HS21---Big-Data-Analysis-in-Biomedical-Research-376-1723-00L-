```python
import pandas as pd
from pybiomart import Server
```

```python
server = Server(host='www.ensembl.org')
```

```python
server.list_marts()
```

# Get information on genes

```python
mart = server['ENSEMBL_MART_ENSEMBL']
mart.list_datasets()
```

```python
dataset = mart['hsapiens_gene_ensembl']

dataset.list_attributes()
```

## Exercise
Create a dictionary of gene names to Ensembl Gene IDs

```python
genes_for_dict_38 = dataset.query(
                                  attributes=['ensembl_gene_id',
                                              'description',
                                              'external_gene_name'])

dict_gene = {}
for i in genes_for_dict_38.itertuples():
    dict_gene[i._3] = i._1
```

```python
# Example
dict_gene['APOB']
```

## Exercise 
Create a list of all the genes on chromosome 5, with gene start, gene end, gene description

```python
dataset.list_filters().head(10)
```

```python
def genes_in_chromosome_of_interest(chromosome_number):
  chrom_of_interest = dataset.query(filters={'chromosome_name' : str(chromosome_number) },
                                    attributes=['ensembl_gene_id',
                                                'description',
                                                'external_gene_name',
                                                'chromosome_name',
                                                'start_position',
                                                'end_position'])
  chrom_of_interest = chrom_of_interest[chrom_of_interest['Gene name'].notna()]
  return chrom_of_interest
```

```python
genes_in_chromosome_of_interest(5)
```
